```python
from noise import pnoise2, snoise2
import matplotlib.pyplot as plt
import matplotlib.image as img
from matplotlib.pyplot import figure
import math
import numpy
```

```python
w, h = 256, 256
octaves = 2
f = 32.0*octaves
```

```python
grid = []
mini, maxi = math.inf, -math.inf
for i in range(h):
    row = []
    for j in range(w):
        val = snoise2(i/f, j/f, octaves)
        mini = min(mini,val)
        maxi = max(maxi,val)
        row.append(val)
    grid.append(row)

for i in range(h):
    for j in range(w):
        grid[i][j] = (grid[i][j]-mini)/(maxi- mini)
```

```python
mini, maxi
```

```python
plt.imshow(grid, cmap='gray')
plt.show()
```

```python
steps = [(0.4, (0,0,225.0)),(0.8, (0,225.0,0)),(1.0, (225.0,225.0,225.0))]
```

```python
terrain = []
for i in range(h):
    row = []
    for j in range(w):
        val = grid[i][j]
        for step in steps:
            if val <= step[0]:
                row.append(step[1])
    terrain.append(row)
```

```python
plt.imshow(numpy.array(terrain))
plt.show()
```
